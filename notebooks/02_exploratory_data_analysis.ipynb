# HR Exploratory Data Analysis

## Objective

The purpose of this notebook is to perform advanced exploratory data analysis to identify attrition patterns, employee behavior trends, and business insights.

## Import Required Libraries

Import essential Python libraries for data analysis and visualization.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_columns', None)
```

## Load Dataset

Use `pd.read_csv()` to read the HR dataset and store it in a pandas DataFrame for analysis.

```python
df = pd.read_csv('../data/Raw/WA_Fn-UseC_-HR-Employee-Attrition.csv')
```

## Attrition by Job Role

Use a countplot to analyze attrition trends across different job roles.

```python
plt.figure(figsize=(12,6))
sns.countplot(x='JobRole', hue='Attrition', data=df)
plt.title('Attrition by Job Role')
plt.xticks(rotation=90)
plt.show()
```

## Attrition by Marital Status

Use a countplot to analyze attrition distribution based on employee marital status.

```python
sns.countplot(x='MaritalStatus', hue='Attrition', data=df)
plt.title('Attrition by Marital Status')
plt.show()
```

## Overtime Analysis

Use a countplot to analyze the relationship between overtime and employee attrition.

```python
sns.countplot(x='OverTime', hue='Attrition', data=df)
plt.title('Overtime vs Attrition')
plt.show()
```

## Work-Life Balance Analysis

Use a countplot to analyze employee attrition across different work-life balance ratings.

```python
sns.countplot(x='WorkLifeBalance', hue='Attrition', data=df)
plt.title('Work-Life Balance vs Attrition')
plt.show()
```

## Job Satisfaction Analysis

Use a countplot to analyze attrition trends based on job satisfaction levels.

```python
sns.countplot(x='JobSatisfaction', hue='Attrition', data=df)
plt.title('Job Satisfaction vs Attrition')
plt.show()
```

## Years at Company Distribution

Use a histogram to analyze employee tenure distribution within the organization.

```python
plt.figure(figsize=(8,5))
sns.histplot(df['YearsAtCompany'], bins=20)
plt.title('Years at Company Distribution')
plt.show()
```

## Business Insights

The exploratory analysis identified major attrition patterns related to overtime, job roles, work-life balance, and job satisfaction. These insights can help HR teams improve employee retention strategies and workforce planning.
